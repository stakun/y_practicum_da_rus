Кредитный отдел банка хочет разобраться, влияет ли семейное положение и количество детей клиента на факт погашения кредита в срок.

**Цель исследования** — на основе предоставленной банком статистики о платежеспособности клиентов проанализировать влияние таких факторов как семейное положение и количество детей на риск непогашения клиентом кредита в срок.

**Задача исследования** — дать ответы на следующие вопросы:
* Есть ли зависимость между количеством детей и возвратом кредита в срок?
* Есть ли зависимость между семейным положением и возвратом кредита в срок?
* Есть ли зависимость между уровнем дохода и возвратом кредита в срок?
* Как разные цели кредита влияют на его возврат в срок?

Ответы должны быть сопровождены интерпретацией — о чём именно говорит полученный результат.

Результаты исследования будут учтены при построении модели кредитного скоринга — специальной системы, которая оценивает способность потенциального заёмщика вернуть кредит банку.

- `children` — количество детей в семье;
- `days_employed` — общий трудовой стаж в днях;
- `dob_years` — возраст клиента в годах;
- `education` — уровень образования клиента;
- `education_id` — идентификатор уровня образования;
- `family_status` — семейное положение;
- `family_status_id` — идентификатор семейного положения;
- `gender` – пол клиента;
- `income_type` — тип занятости;
- `debt` — имел ли задолженность по возврату кредитов;
- `total_income` — ежемесячный доход;
- `purpose` — цель получения кредита;

```python
# Функция days_employed_info (data) будет собирать в датафрейм количество отрицательных, нулевых и положительных значений,
# минимальные, средние, медианные и максимальные значения из столбца 'days_employed' для каждой категории 'income_type',
# а также количество пропусков и выводить табличку с ними на экран

def days_employed_info(data):

    income_type_1 = [] # Столбец с категорией 'income_type'
    income_type_2 = [] # Столбец c количеством отрицательных значений в 'days_employed' для каждого 'income_type'
    income_type_3 = [] # Столбец c количеством нулевых значений в 'days_employed' для каждого 'income_type'
    income_type_4 = [] # Столбец c количеством положительных значений в 'days_employed' для каждого 'income_type'
    income_type_5 = [] # Столбец c минимальными значениями в 'days_employed' для каждого 'income_type'
    income_type_6 = [] # Столбец cо средними значениями в 'days_employed' для каждого 'income_type'
    income_type_7 = [] # Столбец c медианными значениями в 'days_employed' для каждого 'income_type'
    income_type_8 = [] # Столбец c максимальными значениями в 'days_employed' для каждого 'income_type'
    income_type_9 = [] # Столбец c пропусками в 'days_employed' для каждого 'income_type'    

# В цикле по уникальным значениям 'income_type' заполняем списки для каждого из столбцов    
    
    for income_type in data['income_type'].unique():
        income_type_1.append(income_type)
        income_type_2.append(data[(data['days_employed'] < 0) & (data['income_type'] == income_type)].shape[0])
        income_type_3.append(data[(data['days_employed'] == 0) & (data['income_type'] == income_type)].shape[0])
        income_type_4.append(data[(data['days_employed'] > 0) & (data['income_type'] == income_type)].shape[0])
        income_type_5.append(data[data['income_type'] == income_type]['days_employed'].min())
        income_type_6.append(data[data['income_type'] == income_type]['days_employed'].mean())
        income_type_7.append(data[data['income_type'] == income_type]['days_employed'].median())
        income_type_8.append(data[data['income_type'] == income_type]['days_employed'].max())
        income_type_9.append(data[data['income_type'] == income_type]['days_employed'].isna().sum())
        
# Создаем датафрейм, содержащий результаты обзора данных         
        
    days_employed_info = pd.DataFrame({'income_type': income_type_1,
                                       'negative': income_type_2,
                                       'zero': income_type_3,
                                       'positive': income_type_4,
                                       'min': income_type_5,
                                       'mean': income_type_6,
                                       'median': income_type_7,
                                       'max': income_type_8,
                                       'NaN': income_type_9,
                                       })
# Выводим датафрем на экран 

    display(days_employed_info)
```

```python
# Вызов функции для описания столбца 'days_employed'
days_employed_info (data)

# Определим количество строк, содержащих отрицательные значения в столбце 'days_employed'
print('Количество строк с отрицательными `days_employed`, шт.:', data[data['days_employed'] < 0].shape[0])
print('Доля строк с отрицательными `days_employed` в датасете, %:', (data[data['days_employed'] < 0].shape[0]/len(data)*100))
print('Количество строк с пропусками в `days_employed`, шт.:', data['days_employed'].isna().sum())
```

### Категоризация дохода

Задание 16. На основании диапазонов, указанных ниже, создайте в датафрейме `data` столбец `total_income_category` с категориями:
- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.

</br></br>
Например, кредитополучателю с доходом 25000 нужно назначить категорию `'E'`, а клиенту, получающему 235000, — `'B'`.

#### Создание функции `set_income_category`

```python
# Создаем функцию для категоризации дохода
def set_income_category(row):
    total_income = row['total_income']
    if total_income <= 30000 and total_income >= 0:
        return 'E'
    if total_income <= 50000 and total_income >= 30001:
        return 'D'
    if total_income <= 200000 and total_income >= 50001:
        return 'C'
    if total_income <= 1000000 and total_income >= 200001:
        return 'B'
    if total_income > 1000001:
        return 'A'
```

#### Добавление в датафрейм столбца с категорией дохода

```python
# Добавляем в исходный датафрейм столбец с категориями дохода
data['total_income_category'] = data.apply(set_income_category, axis=1)

# Проверяем результат
data.head()
```

### Обоснование заполнения пропусков медианным значением

Задание 24. Объясните, почему заполнить пропуски медианным значением — лучшее решение для количественных переменных.

Заполнение пропусков медианным значением является одним из распространенных методов обработки пропущенных данных для количественных переменных, и это может быть хорошим решением по нескольким причинам:

1. **Сохранение статистических характеристик**: Медиана является мерой центральной тенденции, которая не чувствительна к выбросам, в отличие, например, от среднего значения (средней). Заполнение пропусков медианным значением позволяет сохранить общую структуру данных и предотвратить искажение статистических характеристик распределения.

2. **Избегание искажений**: Заполнение нулевых значений медианным значением позволяет избежать искажения распределения данных. Если бы мы заполнили нулевые значения, например, средним значением, это могло бы значительно повлиять на среднее значение и сделать его менее информативным.

3. **Простота и интерпретируемость**: Заполнение пропусков медианным значением - простой и интуитивно понятный метод, который не требует сложных вычислений или дополнительных данных. Это облегчает интерпретацию результатов.

4. **Устойчивость к выбросам**: Медиана устойчива к выбросам. Это означает, что даже если в данных есть несколько крайних значений (выбросов), медиана остается стабильной. Это полезно, когда в данных есть выбросы, которые могут исказить среднее значение.

5. **Уменьшение потери информации**: Заполнение пропусков медианным значением позволяет сохранить имеющуюся информацию. При удалении строк с пропущенными данными вы можете потерять другие важные характеристики, которые могут быть полезными для анализа.

</br></br>
Однако важно помнить, что выбор метода заполнения пропусков зависит от конкретного контекста и характера данных. В некоторых ситуациях другие методы, такие как интерполяция, могут быть более подходящими. Также необходимо учитывать, что заполнение медианой может быть не лучшим выбором, если есть сильная корреляция между переменными, и такой подход может исказить анализ.
